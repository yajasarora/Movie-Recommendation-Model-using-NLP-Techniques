TMDB5000 Movie Dataset

```python
import pandas as pd
import numpy as np
import ast #Abstract System Trees
import re  #Regular Expression
import nltk #Natural Language Toolkit
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
file_path='tmdb_5000_credits.csv'
df=pd.read_csv(file_path)
file_path='tmdb_5000_movies.csv'
df1=pd.read_csv(file_path)
```

```python
movies=df.merge(df1, on='title')
```

```python
movies.head()
```

Data Preprocessing/Transformation

```python
movies.info()
```

```python
movies=movies[['movie_id','title','genres','keywords','cast','crew','vote_average','overview']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def top_values(data,n):
  name=list()
     #ast.literal_evalconverts a str of list into a list:
  for i, crew in enumerate(ast.literal_eval(data)):
    if i<n:
      #print(i)
      name.append(crew['name'])
    else:
      break
  return name
```

```python
def return_director(data):
  name=list()
  for i, crew in enumerate(ast.literal_eval(data)):
    if crew['job']=='Director':
      name.append(crew['name'])
      break
  return name
```

```python
movies['cast']=movies['cast'].apply(lambda x: top_values(x,5))
movies['crew']=movies['crew'].apply(lambda x: return_director(x) )
movies['genres']=movies['genres'].apply(lambda x: top_values(x,len(ast.literal_eval(x))))
movies['keywords']=movies['keywords'].apply(lambda x: top_values(x,10))
```

```python
movies.isnull().sum()
```

```python
def trim(data):
  data=[i.replace(" ","") for i in data]
  return data
```

```python
movies['cast']=movies['cast'].apply(lambda x: trim(x))
movies['crew']=movies['crew'].apply(lambda x: trim(x))
movies['keywords']=movies['keywords'].apply(lambda x: trim(x))
movies['genres']=movies['genres'].apply(lambda x: trim(x))
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies['vote_average']=movies['vote_average'].apply(lambda x: list([x]))
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['vote_average']
```

```python
movies=movies[['movie_id','title','tags','genres','vote_average']]
```

```python
movies['tags']=movies['tags'].apply(lambda x: " ".join([str(i) for i in x]))
```

```python
movies['tags']=movies['tags'].apply(lambda x: x.lower())
```

```python
movies.head()
```

**Data Cleaing**

```python
def process_text(text):

  stem=PorterStemmer()

  text=re.sub(r'[^\w\s]', "",text) #removing punctuations

  text=word_tokenize(text) #tokenizing

  text=[i for i in text if i not in stopwords.words('english')] #removing stopwords

  text=[stem.stem(i) for i in text]

  text=" ".join(text)

  return text
```

```python
movies['pro_tags']=movies['tags'].apply(lambda x: process_text(x))
```

```python
unique_tags=movies['pro_tags'].str.cat().split()
```

```python
len(set(unique_tags))
```

```python
cv=CountVectorizer(max_features=20000, stop_words='english')
```

```python
vector=cv.fit_transform(movies['pro_tags']).toarray()
```

```python
import numpy as np

# Set numpy print options to display the entire matrix
np.set_printoptions(threshold=np.inf, linewidth=np.inf)

# Print the matrix
print(vector[0])
```

```python
vector.shape
```

```python
cor=cosine_similarity(vector)
```

```python
cor.shape
```

**recommendation function**

```python
def recommend(movie):
    # Check if the movie exists in the DataFrame
    if movie not in movies['title'].values:
        print(f"Movie named:'{movie}' not found in the dataset.")
        return

    # Find the index of the given movie
    movie_index = movies[movies['title'] == movie].index[0]

    # Get the similarity distances from the correlation matrix
    distances = cor[movie_index]

    # Sort the movies based on similarity scores in descending order
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:20]

    # Print the top recommended movie titles
    underline_start = '\033[4m'
    underline_end = '\033[0m'
    print(f"{underline_start}Top recommendations for '{movie}'{underline_end} :\n")
    for i in movies_list:
        title=movies.iloc[i[0]].title
        genre=movies.iloc[i[0]].genres
        vote=movies.iloc[i[0]].vote_average
        print(f"Title: {title}\n Genre: {','.join(genre)}\n Rating: {str(vote[0])}\n")
```

```python
recommend('The Notebook')
```
